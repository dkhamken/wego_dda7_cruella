```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

get api key and get ready to use it

```python
with open('../data/WeGo.json') as fi:
    credentials = json.load(fi)
```

```python
Authorization = credentials['Authorization']
```

first request from API for route 3. note- use 'hiRes' resultion.

```python
URL = 'https://api.goswift.ly/speed-map/nashville/route/3'

params = {
    'direction': '1',
    'startDate': '07-25-2022',
    'resolution': 'hiRes',
}

headers = {'Authorization': Authorization,
          'Content-Type': 'application/json'}
```

```python
response = requests.get(URL, params = params, headers = headers)
```

```python
response.json()
```

turn into dataframe

```python
result = response.json()
route3_7252022 = pd.DataFrame(result)
route3_7252022
```

```python
len(route3_7252022['data'])
```

```python
route3_7252022['data'][0][0:2]
```
